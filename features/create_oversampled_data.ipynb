```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from preprocess import preprocess_text
from features.vectorizer import Vectorizer
from utils.constants import CATEGORIES, TITLE, ABSTRACT, TEXT_FEATURE
from models.supervised_model import SupervisedModel
import warnings
from tqdm import tqdm

import pickle
warnings.filterwarnings("ignore") 
```

```python
with open("features/count_vectorizer.pkl",'rb') as f:
    cv = pickle.load(f)
cv_y = cv['categories']
cv.drop(columns=['categories'], inplace=True)
```

```python
with open("features/tfidf_vectorizer.pkl",'rb') as f:
    tfidf = pickle.load(f)
tfidf_y = tfidf['categories']
tfidf.drop(columns=['categories'], inplace=True)
```

```python
from imblearn.over_sampling import BorderlineSMOTE
```

```python
sm = BorderlineSMOTE(random_state=42)
X_res, y_res = sm.fit_resample(cv, cv_y)
```

```python
y_res.value_counts()
```

```python
sm2 = BorderlineSMOTE(random_state=42)
X_tfidf, y_tfidf = sm2.fit_resample(tfidf, tfidf_y)
```

```python
X_tfidf
```

```python
X_res['categories'] = y_res
X_tfidf['categories'] = y_tfidf
```

```python
X_res.to_pickle("features/oversampled_count_vectorizer.pkl")
X_tfidf.to_pickle("features/oversampled_tfidf_vectorizer.pkl")
```
